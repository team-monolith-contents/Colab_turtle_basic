터틀이 앞으로 이동하고 회전할 수 있어야합니다. 

- t.forward(n) : n 픽셀만큼 전진
- t.backward(n) : n 픽셀만큼 후진
- t.left(n) : n도 만큼 왼쪽으로 회전
- t.right(n) : n도 만큼 오른쪽으로 회전

```python
# turtleScreen을 초기 상태(흰색 배경, 배경 이미지 없음)로 설정합니다. 
clearscreen()
# turtleScreen의 크기를 설정할 수 있습니다. setup(width=픽셀, height=픽셀)
setup(500, 500)

# 그림을 그려줄 거북이를 생성합니다. 
t = Turtle()

# 거북이의 모양을 바꿔요. 
t.shape('turtle')

# 1층
t.penup()
t.goto(-100, -50)
t.pendown()

for i in range(2):
    t.forward(200)
    t.left(90)
    t.forward(50)
    t.left(90)

# 2층
t.penup()
t.goto(-75, 0)
t.pendown()

for i in range(2):
    t.forward(150)
    t.left(90)
    t.forward(50)
    t.left(90)

# 3층
t.penup()
t.goto(-50, 50)
t.pendown()

for i in range(2):
    t.forward(100)
    t.left(90)
    t.forward(50)
    t.left(90)

# 마무리
t.penup()
t.goto(0, 100)
t.pendown()

t.left(90)
t.forward(50)
```
